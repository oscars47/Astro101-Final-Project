## ASAS-SN

```python
asassn_data_dir = '/Volumes/PHLUID/astro/'
```

```python
asassn_summary = pd.read_csv(os.path.join(asassn_data_dir, 'g_band', 'asassn_variables_x.csv'))
asassn_summary.head(10)
```

```python
# find unique types of variables
var_unique = asassn_summary['ML_classification'].unique()
print(len(var_unique))
print(var_unique)
```

```python
asassn_summary_v = pd.read_csv(os.path.join(asassn_data_dir, 'Stars 2022-11-07 04_45.csv'))
asassn_summary_v.head(10)
```

```python
# get unique sources in g and v
unique_sources_g = list(asassn_summary['ID'].unique())
unique_sources_v = list(asassn_summary_v['asassn_name'].unique())
print('unique sources g', len(unique_sources_g))
print('unique sources v', len(unique_sources_v))
unique_sources_merged = list(set(unique_sources_g + unique_sources_v))
print('merged length', len(unique_sources_merged))
```

```python
# get all lc dat files
lc_path = os.path.join(asassn_data_dir, 'g_band/g_band_lcs')
lc_paths = os.listdir(lc_path)
# to get the names, first split by '.dat' and remove final element (.dat), recombine; then split based on '_' and replace with ' '
lc_names = []
for lc in lc_paths:
    temp_list=[]
    temp_name = ''
    temp_list = lc.split('.dat')
    temp_name = temp_list[0]
    temp_list = temp_name.split('_')
    temp_name = temp_list[0] +' ' + temp_list[1]
    lc_names.append(temp_name)

print(len(lc_names))
```

```python
print(unique_sources_g[:100])
print('----------')
print(lc_names[:100])
```

```python
# merge the name lists
lc_merged = list(set(unique_sources_g + lc_names))
print(len(lc_merged))
```
